# 4874. Forth
Forth라는 컴퓨터 언어는 스택 연산을 기반으로 하고 있어 후위 표기법을 사용한다. 예를 들어 3+4는 다음과 같이 표기한다.

3 4 + .

Forth에서는 동작은 다음과 같다.
 

숫자는 스택에 넣는다.

연산자를 만나면 스택의 숫자 두 개를 꺼내 더하고 결과를 다시 스택에 넣는다.

‘.’은 스택에서 숫자를 꺼내 출력한다
 
Forth 코드의 연산 결과를 출력하는 프로그램을 만드시오. 만약 형식이 잘못되어 연산이 불가능한 경우 ‘error’를 출력한다.

다음은 Forth 연산의 예이다.

```python
Forth=['+','-','*','/']

def error_test(c):
    if len(stack)<2:
        return -1
    elif c=='/':
        if stack[-1]==0:
            return -1
        
def cal(c):
    if c=='+':
        return stack.pop(-2)+stack.pop(-1)
    elif c=='-':
        return stack.pop(-2)-stack.pop(-1)
    elif c=='*':
        return stack.pop(-2)*stack.pop(-1)
    elif c=='/':
        return stack.pop(-2)/stack.pop(-1)
    


T = int(input())
# 여러개의 테스트 케이스가 주어지므로, 각각을 처리합니다.
for test_case in range(1, T + 1):
    # ///////////////////////////////////////////////////////////////////////////////////
        result=0
        stack=[]
        L=list(map(str, input().split()))

        print(f"#{test_case}",end=' ')  

        for i in L:
            if i=='.':
                
                if len(stack)==1:
                    result=stack.pop(-1)
                    print(result,end='')

                else:
                    result=-1
                    print("error",end='')
                    break          

            elif Forth.count(i)>0:

                if error_test(i)==-1:
                    result=-1
                    print("error",end='')
                    break
                else:
                    stack.append(int(cal(i)))

            else:
                stack.append(int(i))
                
                
            
        if result==0:
            print("error")
        
        print('')        
    
    # ///////////////////////////////////////////////////////////////////////////////////
```

# 4875. 미로

다음은 5x5 미로의 예이다.

10021

```python
def find_2(l):
    for i_ind,i in enumerate(l):
        for j_ind,j in enumerate(i):
            if j==2:
                return (i_ind,j_ind)
           
           
def move(maze):
    
    for i in direction:
        result=[]
        t_maze=copy2(maze)
        ind=find_2(t_maze)
        t_i=[ind[0]+i[0],ind[1]+i[1]]
        r_list=[]
        if (0<=t_i[0]<n)&(0<=t_i[1]<n):
            if t_maze[t_i[0]][t_i[1]]==1:
                r_list.append(i)
            elif t_maze[t_i[0]][t_i[1]]==3:
                return 3
            elif t_maze[t_i[0]][t_i[1]]==0:
                t_maze[t_i[0]][t_i[1]]=2
                t_maze[ind[0]][ind[1]]=1
                m=move(t_maze)
                if m==3:
                    return 3
                else:
                    continue
                
                
        else:
            r_list.append(i)
        if len(r_list)==4:
            return -1
    return 0

def copy2(maze):
    new_maze=[]
    for i in maze:
        new_maze.append(i.copy())
    return new_maze
            
            
T = int(input())
# 여러개의 테스트 케이스가 주어지므로, 각각을 처리합니다.
for test_case in range(1, T + 1):
    # ///////////////////////////////////////////////////////////////////////////////////
        n=int(input())
        maze=[]
        for i in range(n):
            a=input()
            temp=[]
            for j in range(n):
                temp.append(int(a[j]))
            maze.append(temp)
        direction=[[1,0],[0,1],[-1,0],[0,-1]] 
        done_result=move(maze)
        if done_result==3:
            print(f"#{test_case} 1")
        else:
            print(f"#{test_case} 0")
        
    # ///////////////////////////////////////////////////////////////////////////////////
```

# 4880. 토너먼트 카드게임
사다리 게임이 지겨워진 알고리즘 반 학생들이 새로운 게임을 만들었다. 가위바위보가 그려진 카드를 이용해 토너먼트로 한 명을 뽑는 것이다. 게임 룰은 다음과 같다.
 
1번부터 N번까지 N명의 학생이 N장의 카드를 나눠 갖는다. 전체를 두 개의 그룹으로 나누고, 그룹의 승자끼리 카드를 비교해서 이긴 사람이 최종 승자가 된다.

그룹의 승자는 그룹 내부를 다시 두 그룹으로 나눠 뽑는데, i번부터 j번까지 속한 그룹은 파이썬 연산으로 다음처럼 두개로 나눈다.

두 그룹이 각각 1명이 되면 양 쪽의 카드를 비교해 승자를 가리고, 다시 더 큰 그룹의 승자를 뽑는 방식이다.

다음은 4명이 카드를 비교하는 경우로, 숫자 1은 가위, 2는 바위, 3은 보를 나타낸다. 만약 같은 카드인 경우 편의상 번호가 작은 쪽을 승자로 하고, 처음 선택한 카드는 바꾸지 않는다.


N명이 학생들이 카드를 골랐을 때 1등을 찾는 프로그램을 만드시오.

```python
def div(L):
    l = len(L)//2
    i = len(L)%2
    if l>2:
        return div(L[:l+i])+div(L[l+i:])
    elif l==2:
        
        if i==1:
            return div(L[:l+i])+[L[l+i:]]
        else:
            return [L[:l+i]]+[L[l+i:]]
    else:
        return [L[:l+i]]+[L[l+i:]]
    
def rsp(a,b):
    l=[a,b]
    if abs(l[1]-l[0])==2:
        return l.index(min(l))
    else:
        return l.index(max(l))
    

T = int(input())
for test_case in range(1,T+1):
    n = int(input())
    L = list(map(int,input().split()))
    temp = list([(i+1,item) for i, item in enumerate(L)])

    while True:
        if len(temp)==2:
            break
        new_temp=div(temp).copy()
        for ind, i in enumerate(div(temp)):
            if len(i)==1:
                new_temp[ind] = i[0]
            else:
                new_temp[ind] = i[rsp(i[0][1],i[1][1])]
        temp = new_temp

    print(f"#{test_case} {temp[rsp(temp[0][1],temp[1][1])][0]}")
```

# 4881. 배열 최소 합
NxN 배열에 숫자가 들어있다. 한 줄에서 하나씩 N개의 숫자를 골라 합이 최소가 되도록 하려고 한다. 단, 세로로 같은 줄에서 두 개 이상의 숫자를 고를 수 없다.

조건에 맞게 숫자를 골랐을 때의 최소 합을 출력하는 프로그램을 만드시오.
 

예를 들어 다음과 같이 배열이 주어진다.
 

이경우 1, 5, 2를 고르면 합이 8로 최소가 된다.

```python
def copy2(maze):
    new_maze=[]
    for i in maze:
        new_maze.append(i.copy())
    return new_maze

def make_sum(L,i,sum_):
    if len(L)==1:
        sum_+=L[0][0]
        sum_list.append(sum_)
    else:
        new_L=copy2(L)
        sum_+=new_L[0][i]
        for d in range(1,len(new_L[0])):
            new_L[d].pop(i)
        new_L.pop(new_L.index(L[0]))
        if sum_<min(sum_list):
            for j in range(len(new_L[0])):
                make_sum(new_L,j,sum_)
                
def t_sum(L):
    temp_sum=0
    t_l=copy2(L)
    for i in range(len(L)):
        x=min(t_l[i])
        ind_x=t_l[i].index(x)
        temp_sum+=x
        for j in range(i,n):
            t_l[j].pop(ind_x)
    return temp_sum
                
T = int(input())
# 여러개의 테스트 케이스가 주어지므로, 각각을 처리합니다.
for test_case in range(1, T + 1):
    # ///////////////////////////////////////////////////////////////////////////////////

        n = int(input())
        m=[]
        for i in range(n):
            m.append(list(map(int, input().split()) ))
            
        sum_list=[]
                       
        sum_list.append(t_sum(m))
        
        for i in range(len(m[0])):
            make_sum(m,i,0)


        print(f"#{test_case} {min(sum_list)}")

    # ///////////////////////////////////////////////////////////////////////////////////
```
